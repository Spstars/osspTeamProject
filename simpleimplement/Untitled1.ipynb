```python
ratingdata=[]
ratingdatapostive=[]
ratingdatanegative=[]
with open("ratings.dat", 'r') as file:
    for line in file.readlines():
        arr = line.replace('\n', '').split('::')
        user, item, rating, timestamp = int(arr[0]), int(arr[1]), int(arr[2]), int(arr[3])
        if rating>4.0:
            rating=1
        else:
            rating=0
        ratingdata.append([[user],[timestamp,item]])
        if rating==1:
            ratingdatapostive.append([[user],[timestamp,item]])
        else:
            ratingdatanegative.append([[user],[timestamp,item]])
```

```python
mkgroup=[]
for i in range(6041):
    mkgroup.append([[i],[],[]])
for rating in ratingdata:
    user,time,mv=rating[0][0],rating[1][0],rating[1][1]
    mkgroup[user][1].append([time,mv])
for rating in ratingdatapostive:
    user,time,mv=rating[0][0],rating[1][0],rating[1][1]
    mkgroup[user][2].append([time,mv])
    
for i in range(6040):
    mkgroup[i+1][1].sort(key=lambda x:x[0])
    mkgroup[i+1][2].sort(key=lambda x:x[0])
print(mkgroup[0])
```
